##**Import Dataset**

```python
df = pd.read_csv("https://github.com/YBIFoundation/ProjectHub-MachineLearning/raw/main/Women%20Clothing%20E-Commerce%20Review.csv")
```

<div class="markdown-google-sans">

##**Missing Values**
</div>

**Remove missing values in Reviews columns with No Review text**

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review", inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

<div class="markdown-google-sans">

##**Define Target (y) and Feature (X)**
</div>

```python
df.columns
```

```python
x=df['Review']
```

```python
y=df['Rating']
```

```python
df['Rating'].value_counts()
```

<div class="markdown-google-sans">

##**Train Test Split**
</div>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, stratify= y, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

<div class="markdown-google-sans">

##**Get Feature Text Conversion to Tokens** </div>

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(lowercase = True, analyzer='word', ngram_range=(2, 3), stop_words= 'english', max_features=5000)
```

```python
x_train = cv.fit_transform(x_train)
```

```python
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test = cv.fit_transform(x_test)
```

```python
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

<div class="markdown-google-sans">

##**Get Model Train** </div>

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train, y_train)
```

<div class="markdown-google-sans">

##**Get Model Prediction** </div>

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

<div class="markdown-google-sans">

##**Get Probability of Each Predicted Class** </div>

```python
model.predict_proba(x_test)
```

<div class="markdown-google-sans">

##**Get Model Evaluation** </div>

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
